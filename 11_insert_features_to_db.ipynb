## Connect to database

```python
import os
import psycopg2

# Database connection
conn = psycopg2.connect(
    dbname="deeplabcut_db",
    user="postgres",
    password="1234",
    host="localhost",
    port="5432"
)
cursor = conn.cursor()
```

## Insert data into columns

```python
# Folders
video_dir = r"C:\DeepLabCutProjects\DLC-Atanu-2024-12-25\Videos-to-analyze"
csv_dir = r"C:\DeepLabCutProjects\DLC-Atanu-2024-12-25\Analyzed-videos-filtered"

# Get all filtered csv files
filtered_csvs = [f for f in os.listdir(csv_dir) if f.endswith('_filtered.csv')]

for video_file in os.listdir(video_dir):
    if video_file.endswith(".mp4"):
        base_name = os.path.splitext(video_file)[0]
        video_path = os.path.join(video_dir, video_file)

        # Match CSV using your logic
        matching_csv = [csv for csv in filtered_csvs if base_name in csv]
        if not matching_csv:
            print(f"❌ No matching CSV for: {video_file}")
            continue

        csv_file = matching_csv[0]
        coord_path = os.path.join(csv_dir, csv_file)

        # Check if already inserted
        cursor.execute("SELECT 1 FROM dlc_files WHERE video_path = %s", (video_path,))
        exists = cursor.fetchone()
        if exists:
            print(f"⚠️ Already exists in DB: {video_file}")
            continue

        # Insert
        cursor.execute(
            "INSERT INTO dlc_files (video_path, coord_path) VALUES (%s, %s)",
            (video_path, coord_path)
        )
        print(f"✅ Inserted: {video_file} with CSV {csv_file}")

conn.commit()
cursor.close()
print("✅ All new pairs uploaded.")
```
